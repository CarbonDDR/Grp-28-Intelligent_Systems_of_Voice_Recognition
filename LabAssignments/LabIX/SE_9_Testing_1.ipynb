```python
import datetime
def previous(date):
    try:
        date = datetime.datetime(date["year"], date['month'], date['day'])
    except:
        return "Invalid"
    def test(date2):
        if not (1 <= date2.month <= 12 and 1 <= date2.day <= 31 and 1900 <= date2.year <= 2015):
            return False
        return True

    if test(date):
        return date - datetime.timedelta(days=1)
    else:
        return "Invalid"
```


<h1>Day

1) day between 1 to 31 (both inclusive) valid

2) day more than 31 (31 excluded) invalid

3) day less than 1(1 not included) invalid

<h1>Month

4) month between 1 to 12 (both inclusive) valid

5) month more than 12 (12 excluded) invalid

6) month less than 1 (1 excluded) invalid

<h1>Year

7) year between 1900 to 2015 (both inclusive) valid

8) year more than 2015 (2015 excluded) invalid

9) year less than 1900 (1900 excluded) invalid


```python
test0 = {
    "month": 1,
    "day": 1,
    "year": 1900,
}

previous(test0)
```

```python
import pandas as pd
df = pd.read_csv('/content/test_cases.csv', header=None)
df2 = pd.DataFrame()
df2['month'] = df[0]
df2['day'] = df[1]
df2['year'] = df[2]
df2['Invalid'] = df[3]
df2.head(20)
test_query_obj = {
    "day": None,
    "month": None,
    "year": None
}
adj_mat = [
           [0, 0],
           [0, 0],
]
incorrect_11 = 0
for i in range(len(df2.day)):
    test_i = df2.iloc[i]
    test_query_obj["day"] = test_i.day
    test_query_obj["month"] = test_i.month
    test_query_obj["year"] = test_i.year
    ans = previous(test_query_obj)
    if ans == "Invalid" and test_i.Invalid == "Invalid":
        print("Correct 1", i, ans, test_i, test_query_obj, end="\n\n", sep="\n")
        adj_mat[0][0] += 1
        continue
    if test_i.Invalid == "Invalid":
        print("Problem 1", i, ans, test_i, test_query_obj, end="\n\n", sep="\n")
        adj_mat[0][1] += 1
        continue
    if test_i.Invalid != "Invalid" and ans == "Invalid":
        print("Problem 2", i, ans, test_i, test_query_obj, end="\n\n", sep="\n")
        adj_mat[1][0] += 1
        continue
    else:
        print("Correct 2", i, ans, test_i, test_query_obj, end="\n\n", sep="\n")
        adj_mat[1][1] += 1
        temp = datetime.datetime(
            test_query_obj["year"], 
            test_query_obj['month'], 
            test_query_obj['day']
            )
        if ans != (temp - datetime.timedelta(days=1)):
            print("False!!!!!!")
            incorrect_11 += 1
        continue
adj_mat, incorrect_11
```
